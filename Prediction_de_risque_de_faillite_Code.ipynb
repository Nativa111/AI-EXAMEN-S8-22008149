```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
import shap

# Configuration graphique
warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("Set2")

# ==========================================
# 1. EXPLORATION ET PRÉPARATION DES DONNÉES
# ==========================================

# Chargement du dataset
df = pd.read_csv('Projet-ia- dataset.csv')

print("--- Informations sur le Dataset ---")
df.info()

# Séparation des features (X) et de la cible (y)
X = df.drop('Statut', axis=1)
y = df['Statut']

# Standardisation des données (moyenne nulle et variance unitaire)
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# ==========================================
# 2. CLUSTERING (NON SUPERVISÉ)
# ==========================================

# Réduction de dimension via PCA pour la visualisation
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# K-Means (3 clusters)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters_kmeans = kmeans.fit_predict(X_scaled)

# Clustering Hiérarchique
agglo = AgglomerativeClustering(n_clusters=3)
clusters_agglo = agglo.fit_predict(X_scaled)

# DBSCAN
dbscan = DBSCAN(eps=1.5, min_samples=10)
clusters_dbscan = dbscan.fit_predict(X_scaled)

print(f"\nSilhouette Score (K-Means) : {silhouette_score(X_scaled, clusters_kmeans):.3f}")

# Visualisation des clusters K-Means en 2D
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters_kmeans, palette='viridis', s=60, alpha=0.8)
plt.title('Visualisation des Clusters (K-Means) via PCA')
plt.xlabel('Composante Principale 1')
plt.ylabel('Composante Principale 2')
plt.show()

# ==========================================
# 3. MODÉLISATION SUPERVISÉE (MACHINE LEARNING)
# ==========================================

# Encodage de la variable cible ('En difficulté', 'En faillite', 'Sain')
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Séparation Train (75%) / Test (25%)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.25, random_state=42, stratify=y_encoded
)

# Gestion du déséquilibre avec SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Initialisation des modèles ML
models = {
    "Régression Logistique": LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced'),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)
}

results = {}

for name, model in models.items():
    # Entraînement
    model.fit(X_train_resampled, y_train_resampled)

    # Prédictions sur le test set
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc

    print(f"\n--- Modèle : {name} ---")
    print(f"Accuracy : {acc:.4f}")
    print(classification_report(y_test, y_pred, target_names=le.classes_))

# ==========================================
# 4. COMPARAISON AVEC LE SCORE Z D'ALTMAN
# ==========================================

# Formule proxy du Z-Score (Adaptée aux PME avec nos variables)
df['Altman_Proxy_Z'] = (0.717 * X_scaled['X1_Liquidite'] +
                        0.847 * X_scaled['X3_Rentabilite'] +
                        3.107 * X_scaled['X2_Solvabilite'] +
                        0.420 * X_scaled['X5_Rotation'] -
                        1.0 * X_scaled['X4_Levier'])

def classify_z_score(z):
    if z < -0.5:
        return 'En faillite'
    elif z < 0.5:
        return 'En difficulté'
    else:
        return 'Sain'

df['Prediction_Altman'] = df['Altman_Proxy_Z'].apply(classify_z_score)

altman_accuracy = accuracy_score(df['Statut'], df['Prediction_Altman'])
print(f"\nAccuracy du modèle proxy Altman Z-Score : {altman_accuracy:.4f}")

# ==========================================
# 5. EXPLICABILITÉ (SHAP) SUR RANDOM FOREST
# ==========================================

best_model = models['Random Forest']
explainer = shap.TreeExplainer(best_model)
shap_values = explainer.shap_values(X_test)

# Identification de l'index de la classe "En faillite"
index_faillite = list(le.classes_).index('En faillite')

print("\n--- Analyse de l'impact des variables (SHAP) ---")
# Visualisation globale de l'importance des "drivers" de risque
shap.summary_plot(shap_values[:, :, index_faillite], X_test, plot_type="bar", feature_names=X.columns)

# Visualisation détaillée avec la direction de l'impact
shap.summary_plot(shap_values[:, :, index_faillite], X_test, feature_names=X.columns)
```
